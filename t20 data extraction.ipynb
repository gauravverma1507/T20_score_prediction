```python
import numpy as np
import pandas as pd
import os
from tqdm import tqdm # show the processinng speed#Total Quality Data Management 
from yaml import safe_load # yaml se data pandas me ajayga
```

```python
#os modules ki help se yaml files ka naam only extract kara hai ; filenames me 
filenames =[]
for file in os.listdir('data'):
    filenames.append(os.path.join('data',file))
```

```python
# imp. code
#har file me ghus ke normalize kar rhe hai or safe_load ki vajah se  yaml files handle ho rhi hai
#data ka backup bhi hona jaroori hai to hum kya karenge hum backup naam se ek variabble lenge or usme sab store kar denge
#create a dataframe
#tqdm se data progress show hoti hai 

final_df = pd.DataFrame()
counter=1
for file in tqdm(filenames):          
    with open(file,'r') as f:
        df = pd.json_normalize(safe_load(f))
        df['match_id']= counter
        final_df = final_df.append(df)
        counter+=1
        
#error at last is due to inconsistency in yaml file        
```

```python
matches = pickle.load(open('dataset_level1.pkl','rb'))
matches
```

```python
# we get every ball data 
# iloc chooses 1st row iloc[0]

matches.iloc[0]['innings'][0]['1st innings']['deliveries']
```

```python
# this code will seperate 963 matches balls by balls with other prior information also 


count = 1 
delivery_df = pd.DataFrame()
for index, row in matches.iterrows():
    if count in [75,108,150,180,268,360,443,458,584,748,982,1052,1111,1226,1345]:
        count+=1
        continue
    count+=1
    
    ball_of_match=[]
    batsman=[]
    bowler=[]
    runs=[]
    player_of_dismissed = []
    teams=[]
    batting_team=[]
    match_id=[]
    city=[]
    venue=[]
    
    for ball in row['innings'][0]['1st innings']['deliveries']:
        for key in ball.keys():
            match_id.append(count)
            batting_team.append(row['innings'][0]['1st innings']['team'])
            teams.append(row['info.teams'])
            ball_of_match.append(key)
            batsman.append(ball[key]['batsman'])
            bowler.append(ball[key]['bowler'])
            runs.append(ball[key]['runs']['total'])
            city.append(row['info.city'])
            venue.append(row['info.venue'])
            
            try:
                player_of_dismissed.append(ball[key]['wicket']['player_out'])
            except:
                player_of_dismissed.append('0')
                    
    loop_df = pd.DataFrame({
        
        'match_id':match_id,
        'teams':teams,
        'batting_team': batting_team,
        'ball':ball_of_match,
        'batsman': batsman,
        'bowler':bowler,
        'runs': runs,
        'player_dismissed':player_of_dismissed,
        'city':city,
        'venue':venue,       
        
    })
    
    delivery_df = delivery_df.append(loop_df)
```

```python
delivery_df
```

```python
#we need some features from this data set 
# we need
'''
batting,bowling team
current score
wickets left
run rate
city
balls left
last_5 over runs
'''
# start finding
# we know we have batting team but we dont have bowling team 
```

```python
# for bowling team

def bowl(row):
    for team in row['teams']:
        if team != row['batting_team']:
            return team
        
# here logic is teams column consist 2 team and batting_team is 1 
# not in batting team will be bowling_team 
```

```python
delivery_df['bowling_team']= delivery_df.apply(bowl,axis=1)
```

```python
# we will take only top 10 teams who have played a number of matches because if we takes a lot of number of teams which have low matches they will generate bias in the model 
teams = {
    'Australia',
    'India',
    'Bangladesh',
    'New Zealand',
    'South Africa',
    'England',
    'West Indies',
    'Afghanistan',
    'Pakistan',
    'Sri Lanka'
}
```

```python
delivery_df = delivery_df[delivery_df['batting_team'].isin(teams)]
delivery_df = delivery_df[delivery_df['bowling_team'].isin(teams)]
```
